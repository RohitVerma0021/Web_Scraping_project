```python
#Now Import necessary library
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np

#First create the empty list 
Reviews=[]
Rating=[]
Headquater=[]
No_of_Employee=[]
Comapny_Old=[]
Company_Type=[]
Comapny_Name=[]


#Now Create Empty DataFrame to store the data 
final_df=pd.DataFrame()


#Start's the project's
i=0
while i<30:
  url="https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}".format(i)
  headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  webpage=requests.get(url,headers=headers).text
  soup=BeautifulSoup(webpage,'lxml')

  #Fetch the tags from inspact
  company=soup.find_all("div",class_="company-content-wrapper")
  i+=1

for i in company:
  try:
    Comapny_Name.append(i.find('h2').text.strip())
  except:
    Comapny_Name.append(np.nan)
  try:
    Rating.append(i.find('p',class_="rating").text.strip())
  except:
    Rating.append(np.nan)
  try:
    Headquater.append(i.find_all("p",class_="infoEntity")[1].text.strip())
  except:
    Headquater.append(np.nan)
  try:
    No_of_Employee.append(i.find_all("p",class_="infoEntity")[3].text.strip())
  except:
    No_of_Employee.append(np.nan)
  try:
    Comapny_Old.append(i.find_all("p",class_="infoEntity")[2].text.strip())
  except:
    Comapny_Old.append(np.nan)
  try:
    Company_Type.append(i.find_all('p',class_="infoEntity")[0].text.strip())
  except:
    Company_Type.append(np.nan)
  try:
    Reviews.append(i.find("a",class_="review-count sbold-Labels").text.strip())
  except:
    Reviews.append(np.nan)
  
df=pd.DataFrame({
    'Comapny_Name':Comapny_Name,
    'Company_Type':Company_Type,
    'Rating':Rating,
    "Headquater":Headquater,
    "No_of_Employee":No_of_Employee,
    "Comapny_Old":Comapny_Old
})
final_df.append(df,ignore_index=True)
```

```python
final_df.to_csv("company_list_name.csv")
```
